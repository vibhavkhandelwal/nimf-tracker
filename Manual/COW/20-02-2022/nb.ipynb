```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv("data.csv", index_col=0, parse_dates=True)
```

```python
ax = df.plot(secondary_y="Real lending rate (CPI+WPI based)", figsize=(8,4.5), x_compat=True)

ax.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))
ax.right_ax.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))

plt.title("Real lending rate and real credit growth are very low!")

ax.grid(visible=False)
plt.grid(visible=False)

ax.set_xlabel("Lending rate taken as base rate and prime lending rate (as and when applicable)")

ax.legend(loc='lower center', bbox_to_anchor=(0.25, -0.5), fancybox=True, ncol=2)
plt.legend(loc='lower center', bbox_to_anchor=(0.75, -0.5), fancybox=True, ncol=2)

# plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)

plt.grid(False)

plt.tight_layout()
plt.savefig("cow20feb.png", dpi=250, bbox_inches='tight')
```
